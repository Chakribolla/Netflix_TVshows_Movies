```python
crude_df = crude_df.drop(['show_id'], axis = 1)
crude_df
```

```python
sns.heatmap(crude_df.isna(),cmap = 'viridis')
```

```python
tvshows = crude_df.loc[crude_df["type"]=="TV Show"]
tvshows = tvshows.reset_index(drop=True)
print("number of tvshows",len(tvshows))
```

```python
movies = crude_df.loc[crude_df["type"]=="Movie"]
movies = movies.reset_index(drop=True)
print("number of movies",len(movies))
```

```python
movies.director.nunique()
```

```python
tvshows.director.nunique()
```

```python
tvshows.director.isna().sum()
```

```python
movies['director'].fillna(movies['director'].mode()[0], inplace = True)
movies
```

```python
tvshows['director'].fillna(tvshows['director'].mode()[0], inplace = True)
tvshows
```

```python
movies['cast'].fillna(movies['cast'].mode()[0], inplace = True)
tvshows['cast'].fillna(tvshows['cast'].mode()[0], inplace = True)
```

```python
movies['duration'] = movies['duration'].apply((lambda x : str(x).split(" ")))
movies['duration'] = movies['duration'].apply(lambda x : x[0])
#movies
```

```python
tvshows['duration'] = tvshows['duration'].apply((lambda x : str(x).split(" ")))
tvshows['duration'] = tvshows['duration'].apply(lambda x : x[0])

#tvshows
```

```python
df = pd.concat([movies,tvshows])
df
```

```python
df.isna().sum()/len(df)*100
```

```python
df.country.fillna(df["country"].mode()[0],inplace=True)
```

```python
    df.dropna(axis =0, how="any", inplace = True)
df
```
